```python
# Drop rows with null values in the winner column (adjust name if needed)
df = df.dropna(subset=['Match_Winner_Id']).copy()

# Forward fill for other columns
df = df.ffill()

# Convert 'date' to datetime format
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
```

```python
df = df.drop('Unnamed: 7', axis=1)
```

```python
# Create a feature for whether the toss winner also won the match
df['toss_win_match_win'] = (df['Toss_Winner_Id'] == df['Match_Winner_Id']).astype(int)

# One-hot encode teams and venue
df_encoded = pd.get_dummies(df, columns=['Team_Name_Id', 'Opponent_Team_Id', 'Venue_Name'], drop_first=True)

# Example: encode toss decision
df_encoded['bat_first'] = (df['Toss_Decision'] == 'bat').astype(int)

# Define target variable
y = df_encoded['Match_Winner_Id']
X = df_encoded.drop(columns=['Match_Winner_Id', 'Match_Id'])  # drop non-predictive columns
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X = df.select_dtypes(include=np.number).drop('Match_Winner_Id', axis=1)
y = LabelEncoder().fit_transform(df['Match_Winner_Id'])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Encode categorical columns before splitting data
from sklearn.preprocessing import LabelEncoder

# Team and Player IDs
for col in ['Team_Name_Id', 'Opponent_Team_Id', 'Player_Id']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

# Match outcome encoding
df['Match_Winner_Id'] = LabelEncoder().fit_transform(df['Match_Winner_Id'])
```

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Define base learners
base_learners = [
    ('lr', LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)),
    ('svc', SVC(probability=True)),  # Needed for stacking
    ('nb', GaussianNB()),
    ('dt', DecisionTreeClassifier())
]

# Define final estimator
final_estimator = RandomForestClassifier()

# Create stacking classifier
# Modified stacking classifier with proper params
stacked_model = StackingClassifier(
    estimators=base_learners,
    final_estimator=final_estimator,
    cv=5,
    n_jobs=-1,
    passthrough=False  # Ensure no raw features pass through
)

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

stacked_model.fit(X_train_imputed, y_train)
y_pred = stacked_model.predict(X_test_imputed)


# Predict
y_pred = stacked_model.predict(X_train)

# Evaluate
accuracy = accuracy_score(y_train, y_pred)
print(f"Training Accuracy: {accuracy:.4f}")
```

```python
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

# XGBoost + sklearn imports
from xgboost import XGBClassifier
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
import joblib

# Load your trained model (after saving it)
# xgb_model = joblib.load("xgb_model.pkl")

# Or: If training here (quick example)
# Train your model somewhere before and assume it's saved/loaded
# xgb_model.fit(X_train_imputed, y_train)

# Example dummy function to simulate prediction
def predict_winner(team1, team2):
    # Replace with actual input preprocessing and model prediction
    sample_input = pd.DataFrame([[team1, team2]], columns=["team1_id", "team2_id"])
    # Convert to numerical, scale, impute, etc. as your real model needs
    return xgb_model.predict(sample_input)[0]  # Output predicted winner ID or label

# --- LangChain Part ---
model = OllamaLLM(model="llama3.2")

# Define a basic prompt template
template = (
    "You are an assistant that helps predict the winner between two IPL teams.\n"
    "User: {question}\nAI:"
)
chat_prompt = ChatPromptTemplate.from_template(template)

def handle_conversation():
    print("Welcome to the IPL Prediction Bot! Type 'exit' to quit.\n")
    
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            break
        
        # Simple custom logic to check for prediction pattern
        if "predict" in user_input.lower() and "vs" in user_input.lower():
            try:
                # Parse team names (simplified logic)
                team1, team2 = [x.strip() for x in user_input.lower().split("vs")]
                prediction = predict_winner(team1, team2)
                print("Bot (ML Prediction):", prediction)
            except Exception as e:
                print("Bot (Error): Could not parse input or run prediction.", e)
        else:
            # Use LLM response
            formatted_prompt = chat_prompt.format(question=user_input)
            result = model.invoke(formatted_prompt)
            print("Bot:", result)

if __name__ == "__main__":
    handle_conversation()
```
